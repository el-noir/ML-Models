```python
import numpy as np
import matplotlib.pyplot as plt

# Data
house_sizes = np.array([850, 900, 1200, 1350, 1500])
house_prices = np.array([180, 190, 250, 275, 300])
```

```python
# Calculate slope (m) and intercept (b) using least squares method
mean_x = np.mean(house_sizes)
mean_y = np.mean(house_prices)

numerator = np.sum((house_sizes - mean_x) * (house_prices - mean_y))
denominator = np.sum((house_sizes - mean_x)**2)
m = numerator / denominator
b = mean_y - m * mean_x

# Calculate the initial regression line predictions
initial_predictions = m * house_sizes + b

# Print the initial slope and intercept
print(f"Initial Slope (m): {m}")
print(f"Initial Intercept (b): {b}")
```

```python
# Initialize parameters for Gradient Descent
learning_rate = 0.0000001
m_gd, b_gd = m, b  # Start with initial slope and intercept
iterations = 5

# Perform Gradient Descent
for i in range(iterations):
    # Calculate predictions
    predictions = m_gd * house_sizes + b_gd
    error = house_prices - predictions

    # Calculate gradients
    m_gradient = -(2 / len(house_sizes)) * np.sum(error * house_sizes)
    b_gradient = -(2 / len(house_sizes)) * np.sum(error)

    # Update slope and intercept
    m_gd -= learning_rate * m_gradient
    b_gd -= learning_rate * b_gradient

    # Print updated values
    print(f"Iteration {i+1}: Slope (m): {m_gd}, Intercept (b): {b_gd}")

# Calculate the final regression line predictions
final_predictions = m_gd * house_sizes + b_gd
```
